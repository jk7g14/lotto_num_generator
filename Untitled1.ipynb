```python
import math
def normal_round(n):
    if n - math.floor(n) < 0.5:
        return math.floor(n)
    return math.ceil(n)
```

```python
def get_max_list(nums_list):
    fq= defaultdict( int )
    for w in nums_list:
        fq[w] += 1
    return [k for k, v in fq.items() if v == max(fq.values())], fq
```

```python
index_list = list(map(normal_round, mul_list))
print(index_list)
```

```python
tmp_df = df[0:index_list[4]]
```

```python
front_nums = tmp_df.iloc[:, list(range(0,6))]
back_nums = tmp_df.iloc[:, [6]]
print(front_nums.size)
print(back_nums.size)
```

```python
back_nums_list = back_nums.values.reshape(1, -1).ravel().tolist()
```

```python
back_max, fq_back = get_max_list(back_nums_list)
```

```python
front_nums_list = front_nums.values.T.ravel().tolist()
```

```python
lotto_num_list = []
max_nums, front_fq = get_max_list(front_nums_list)
for num in max_nums:
    lotto_num_list.append(num)
    del front_fq[num]
print(lotto_num_list)

while len(lotto_num_list) < 5:
    max_nums = [k for k, v in front_fq.items() if v == max(front_fq.values())]
    for num in max_nums:
        del front_fq[num]
        lotto_num_list.append(num)
        print(lotto_num_list)
```
